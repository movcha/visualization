Part 6 Subplots and Combining Visualizations


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,75,50)
x2 = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y2 = np.array([110,180,240,99,220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1,
                               figsize=(7,3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows = 1,
                               figsize=(10,5))
ax1.scatter(x1,y1, color = 'indigo', marker = '*') 
ax2.bar(x2,y2, color='green')

ax1.set_xlabel("X-axis", fontdict={'family': 'serif', 'color': 'green', 'weight': 'normal', 'size': 12})
ax1.set_ylabel("Y-axis", fontdict={'family': 'serif', 'color': 'green', 'weight': 'normal', 'size': 12})
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4))

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                   figsize=(7, 4))


someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
someaxes["ax1"].set_xlabel("A Big Label!",fontsize=18)
someaxes["ax2"].set_xlabel("Another Label",fontsize=18)
someaxes["ax3"].set_xlabel("Label 2: 2 Fast 2 Furious",fontsize=18)
```

```python
x = np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1 = np.array([110,180,240,99,220])
y2 = np.array([170,100,90,120,50])

fig, ax = plt.subplots(figsize=(7,3))

ax.bar(x,y1,
       color='indigo')
ax.plot(x,y2,
        color = 'red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color='red')

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd,
            fmt = "none")
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color='red')

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd,
            fmt = "none",
            ecolor = 'indigo',
            elinewidth = 4,
            capsize = 6,
            capthick =4)
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x,y2,color='red')

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd,
            fmt = "none",
            ecolor = 'indigo',
            elinewidth = 4,
            capsize = 6,
            capthick =4,
            errorevery =2)
```
